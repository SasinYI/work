```python
# Функцкия построения фазового портрета
def draw_f_plot(x, y):
    plt.plot(x, y)
    plt.title("Фазовый портрет")
    plt.xlabel('x')
    plt.ylabel('x`')
    plt.grid()
    plt.show()
```

```python
# Функция построения графика решения
def draw_plot(x, y, t):
    plt.plot(t, x, label = 'x')
    plt.plot(t, y, label = 'x`')
    plt.title("Решение дифференциального уравнения")
    plt.xlabel('t')
    plt.ylabel('x')
    plt.legend()
    plt.grid()
    plt.show()
```
