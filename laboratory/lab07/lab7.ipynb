```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math

t = np.linspace(0,10,3000)
x0 = 16
N = 910
f1 = 0.84
f2 = 0.00002

def g(x,t):
    x = (f1 + f2*x)*(N-x)
    return(x)

x = odeint(g,x0,t)


def g1(x,t):
    x = (f1)*(N-x)
    return(x)

x1 = odeint(g1,x0,t)


def g2(x,t):
    x = (f2*x)*(N-x)
    return(x)

x2 = odeint(g2,x0,t)


plt.plot(t,x,label = "X1(t)")
plt.plot(t,x1,label = "X2(t)")
plt.plot(t,x2,label = "X3(t)")
plt.legend()
plt.show

for i in range(20):
    print(x[i],x1[i],x2[i])
```
