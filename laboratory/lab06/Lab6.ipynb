```python
import numpy as np
import matplotlib.pylab as plt

N = 1120
I0 = 230
R0 = 45
S0 = N - I0 - R0
a = 0.01
b = 0.02
t = np.linspace(0,1,100)

def dI (I):
    I -= b*I
    return(I)

def dR (R):
    R += b*R
    return(R)

I = np.zeros(100)
I[0] = I0
for i in range(len(I) -1):
    I[i+1] = dI(I[i])

R = np.zeros(100)
R[0] = R0
for i in range(len(R) -1):
    R[i+1] = dR(R[i])
    
plt.plot(t,I,label = "I(t)")
plt.plot(t,R,label = "R(t)")
plt.legend()
plt.show()
```

```python
def dS (S):
    S -= a*S
    return(S)

S = np.zeros(100)
S[0] = S0
for i in range(len(S)-1):
    S[i+1] = dS(S[i])

def dI_2(S,I):
    I = I - b*I + a*S
    return(I)
for i in range(len(I)-1):
    I[i+1] = dI_2(I[i],S[i])
    
plt.plot(t,I,label = "I(t)")
plt.plot(t,R,label = "R(t)")
plt.plot(t,S,label = "S(t)")
plt.legend()
plt.show()
```
