```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def a(tau,p):
    a = pcr/(tau**2 * p**2 * N * q)
    return a


def c(tau,p):
    c = (pcr - p)/(tau * p)
    return c
```

```python
def dx(x, t):
    dx1 = x[0] - ((b/c1) * x[0] * x[1]) - ((a1/c1) * x[0]**2)
    dx2 = ((c2/c1)*x[1]) - ((b/c1) * x[0] * x[1]) - ((a2/c1) * x[1]**2)
    return[dx1, dx2]
```

```python
def draw_plot(x, y, t):
    plt.plot(t, x, label = 'средства фирмы 1')
    plt.plot(t, y, label = 'средства фирмы 2')
    plt.title("Решение дифференциального уравнения")
    plt.xlabel('t')
    plt.ylabel('x(t), y(t)')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
v0 = np.array([7.5,8.5])
pcr = 40
N = 95
q = 1
tau1 = 30
tau2 = 27
p1 = 11.5
p2 = 9.5

t = np.linspace(0,10,100)
```

```python
a1 = a(tau1,p1)
a2 = a(tau2,p2)
b = pcr/(tau1**2 * tau2**2 * p2**2 * p1**2 * N * q)
c1 = c(tau1,p1)
c2 = c(tau2,p2)
theta = t/c1
```

```python
x = odeint(dx, v0, theta)


xpoint = [elem[0] for elem in x] 
ypoint = [elem[1] for elem in x]

draw_plot(xpoint,ypoint,theta)
```

```python
x = odeint(dx, v0, theta)


xpoint = [elem[0] for elem in x] 
ypoint = [elem[1] for elem in x]

draw_plot(xpoint,ypoint,theta)
```
