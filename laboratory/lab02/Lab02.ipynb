```python
import numpy as np
from math import sqrt,pi,tan
import scipy.integrate as itg
import matplotlib.pyplot as plt


def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def dr(r, theta):
    dr = r / sqrt(19.25)
    return dr

def dx(t):
    xt = tan(fi) * t
    return xt
```

```python
s = 15.5 # Начальное расстояние от лодки до катера
fi = 3 * pi / 4 

# Для случая 1
# Для катера береговой охраны
r0 = 15.5 / 5.5
theta0  = 0
theta = np.arange(theta0, 2 * pi, 0.01)
# Для лодки браконьеров
t0 = 0
t = np.arange(t0, 13, 1)
```

```python
r = itg.odeint(dr, r0, theta)

[rho, phi] = cart2pol(t, dx(t))

plt.polar(theta, r)
plt.polar(phi, rho)
```
